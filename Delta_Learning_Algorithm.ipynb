# Delta Learning

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([0.5,2.5])
y = np.array([0.2,0.9])
w = np.array([0,0])
b = 0
```

```python
def sigmoid(y_in):
  y_net = 1/(1 + np.exp(-y_in))
  return y_net
```

```python
def d_w(x,y,w,b,alpha):
  y_hat = perceptron(x,w,b)

  return alpha*(y-y_hat)*y_hat*(1-y_hat)*x


def d_b(x,y,w,b,alpha):
  y_hat = perceptron(x,w,b)
  loss_sq = y - y_hat
  return alpha*(y-y_hat)*y_hat*(1-y_hat), loss_sq**2
```

```python
def perceptron(x,w,b):
  y_in = np.dot(w,x) + b
  y_net = sigmoid(y_in)
  return y_net
```

## Gradient Descent

Stochastic GD

```python
plot_w = []
plot_b = []
plot_e = []

def Stochastic_descent(X,Y,w,b,epoch):
  for i in range(epoch):
    alpha = 1
    dw=0
    db=0
    loss_sq = 0
    for x,y in X,Y:
      dw = d_w(x,y,w,b,alpha)
      w = w + dw
      loss_sq, db = d_b(x,y,w,b,alpha)
      db = b + db
      loss_sq+=loss_sq

    plot_b.append(db)
    plot_w.append(w)
    plot_e.append(1/2*(loss_sq))
```

```python
Stochastic_descent(X,y,w,b,1000)
```
